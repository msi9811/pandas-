#데이터 전처리
일별로 교통량과 교통사고의 상관관계를 확인하고 싶음
->각 데이터를 일별로 나타내고 합칠 수 있게 동일화

##교통량

```python
#총교통량이 모든 인덱스에서 오류없이 정상적으로 나타냄을 확인
raw_traffic[(raw_traffic["1종교통량"]+raw_traffic["2종교통량"]+raw_traffic["3종교통량"]+
            raw_traffic["4종교통량"]+raw_traffic["5종교통량"]+raw_traffic["6종교통량"])!=raw_traffic["총교통량"]]
```

```python
fix_traffic=raw_traffic.drop(["1종교통량","2종교통량","3종교통량","4종교통량","5종교통량","6종교통량",],axis=1)
fix_traffic
```

```python
#Unnamed: 12는 빈 컬럼이므로 삭제
fix_traffic=fix_traffic.drop("Unnamed: 12",axis=1)
```

```python
#날짜마다 여러가지 인덱스가 존재 -> 얘네들을 합쳐도 상관없을까? 확인이 필요하다.
fix_traffic[fix_traffic["집계일자"]==20200101]
```

```python
#구분코드는 fix_traffic 앞 5컬럼의 숫자들을 나열시킨 값
#fix_traffic 구분코드의 중복값이 없으므로 인덱스마다의 교통량 중복값은 없다고 생각 -> 날짜끼리 묶어도 상관없다!
fix_traffic["구분코드"]=fix_traffic["집계일자"]*100000+fix_traffic["입출구구분코드"]*10000+fix_traffic["TCS하이패스구분코드"]*1000+fix_traffic["고속도로운영기관구분코드"]*10+fix_traffic["영업형태구분코드"]
len(fix_traffic["구분코드"].unique())
```

```python
fix1_traffic=fix_traffic.drop(["입출구구분코드","TCS하이패스구분코드","고속도로운영기관구분코드","영업형태구분코드","구분코드"],axis=1)
fix1_traffic
```

```python
result_traffic=fix1_traffic.groupby("집계일자").sum()
result_traffic
```

##교통사고

```python
#가장 밑 한 인덱스가 모두 NaN임
raw_accident=raw_accident.dropna()
raw_accident.head(2)
```

교통량 데이터프레임과 형식을 맞추기

```python
temp=raw_accident.drop("도로종류 ",axis=1)
temp=temp.drop(0)
fix_accident=pd.melt(temp,id_vars=["사고월","사고년도"])
fix_accident.columns=["사고월","분류","사고일","값"]
```

```python
fix_accident
```

```python
#사고일 컬럼의 2020은 월 합산량이므로 이번 분석에 필요가 없음
fix_accident=fix_accident.drop(fix_accident[fix_accident["사고일"]=="2020"].index)
```

인덱스에 들어갈 양식 맞추기

```python
def date(x):
  result=x.split(".")[1]
  if len(result)==1:
    result="0"+result
  return result
```

```python
def month(x):
  result=x.split("월")[0]
  return result
```

```python
fix_accident["일"]=fix_accident["사고일"].apply(date)
fix_accident["월"]=fix_accident["사고월"].apply(month)
fix_accident["집계일자"]="2020"+fix_accident["월"]+fix_accident["일"]
fix_accident["집계일자"]=pd.to_numeric(fix_accident["집계일자"])
fix_accident.index=fix_accident["집계일자"]
```

```python
fix_accident.info()
```

"값" 컬럼 확인

```python
#"-"가 포함되어 있음
#처음엔 0의 다른 표현으로 생각했지만 0도 값에 존재
#뭘까?
fix_accident["값"].unique()
```

```python
#"-"는 2020년에 존재하지 않는 날짜임을 확인 -> 제거/ 이후 int값으로
fix_accident=fix_accident.drop(fix_accident[fix_accident["값"]=="-"].index)
fix_accident["값"]=pd.to_numeric(fix_accident["값"])
```

```python
fix_accident
```

최종처리

```python
#필요없는 컬럼 제거
result_accident=fix_accident.drop(["사고월","사고일","일","월","집계일자"],axis=1)
```

```python
#날짜순 정렬
result_accident=result_accident.sort_index()
```

```python
result_accident
```

#데이터 병합

교통사고 데이터를 사고/사망/중상/부상으로 분류된 데이터를 분리

```python
cnt_accident=result_accident[result_accident["분류"]=="사고건수"]
cnt_accident=cnt_accident.drop("분류",axis=1)
cnt_accident.columns=["사고건수"]
```

```python
cnt_death=result_accident[result_accident["분류"]=="사망자수"]
cnt_death=cnt_death.drop("분류",axis=1)
cnt_death.columns=["사망자수"]
```

```python
cnt_injury=result_accident[result_accident["분류"]=="부상자수"]
cnt_injury=cnt_injury.drop("분류",axis=1)
cnt_injury.columns=["부상자수"]
```

```python
cnt_serious=result_accident[result_accident["분류"]=="(중상자수)"]
cnt_serious=cnt_serious.drop("분류",axis=1)
cnt_serious.columns=["중상자수"]
```

데이터 병합

```python
result_data=pd.concat([result_traffic, cnt_accident, cnt_death, cnt_injury, cnt_serious], axis=1)
result_data
```

```python
result_data["교통량대비사고건수"]=result_data["사고건수"]/result_data["총교통량"]*100000000
result_data["교통량대비부상자수"]=result_data["부상자수"]/result_data["총교통량"]*100000000
result_data["교통량대비중상자수"]=result_data["중상자수"]/result_data["총교통량"]*100000000
result_data
```
